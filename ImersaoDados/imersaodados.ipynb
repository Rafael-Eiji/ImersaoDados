```python
#Verifica se tem algum valor numérico na terceira coluna
dados['Valor'].str.split(expand = True)[2].unique()
```

```python
#Adiciona os valores na tabela
dados[['Moeda', 'Valor_anuncio', 'Tipo_anuncio']] = dados['Valor'].str.split(expand = True)
dados
```

```python
#Filtra apenas pelos dados que não são alugueis
dados_vendas = dados[dados['Tipo_anuncio'].isnull()]
dados_vendas
```

```python
dados_vendas['Tipo_anuncio'].unique()
```

```python
dados_vendas.info()
```

```python
dados_vendas["Valor_anuncio"] = dados_vendas["Valor_anuncio"].str.replace(".", "").astype('float64')
```

```python
dados_vendas
```

```python
dados_vendas.groupby('Bairro').mean().plot.hist()
```

```python
pd.set_option("display.precision", 2)
pd.set_option("display.float_format", lambda x:"%.2f" %x)
dados_vendas.describe()
```

```python
dados_vendas['Valor_em_milhoes'] = dados_vendas['Valor_anuncio']/ 1000000
dados_vendas
```

```python
dados_vendas.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fonte = {"size": 15}
sns.set_style("whitegrid")
sns.set()
plt.figure(figsize=(10 , 8))
paleta = list(sns.color_palette("mako"))
ax = sns.histplot(data=dados_vendas, x = 'Valor_em_milhoes', kde=True, palette=paleta)
ax.set_title("Histograma valor imóveis")
plt.title("Número de casas pelo preço de venda", fontdict = fonte)
plt.xlabel("Preço das casas em milhões", fontdict = fonte)
plt.ylabel("Número de casas", fontdict = fonte)
plt.xlim((-50/1000000, 10))
plt.show()
```

## Desafio Aula 2

- Preço do metro quadrado por bairro e plotar em um gráfico ideal
- Explorar as bibliotecas de visualizações e colocar as suas conclusão
- Pesquisar um visualização para analisar os quartis, mediana e outliers

```python
media_dados = dados_vendas.groupby('Bairro').mean()
media_dados['Valor/m2'] = media_dados['Valor_anuncio']/media_dados['Metragem']
media_m2 = media_dados.sort_values(by= 'Valor/m2',ascending=  False).head(10)
media_m2
```

```python
fonte = {"size": 20}
plt.figure(figsize=(20 , 10))
sns.color_palette("light:#5A9", as_cmap=True)
ax = sns.barplot(data = media_m2, x= media_m2.index , y = 'Valor/m2', palette="coolwarm")
plt.title("Média de Preços/m² por bairro", fontdict = fonte)
plt.xlabel("Bairro", fontdict = fonte)
plt.ylabel("Valor por m²", fontdict = fonte)
```

```python
plt.figure(figsize=(20 , 10))
sns.boxplot(y = dados_vendas['Valor_em_milhoes'])
```

```python
sns.set()
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
plt.figure(figsize=(15, 8))
ax = sns.histplot(data = dados_vendas, x = "Valor_em_milhoes", kde = False, color = "skyblue", lw=2, bins=65, edgecolor="black")
ax.set_title("Imóveis em São Paulo", size=16)
ax.set(xlabel='Valor do imóvel em milhões de reais', ylabel='Quantidade de imóveis à venda')
plt.xlim((-0.5, 10.5))
plt.show()
```
